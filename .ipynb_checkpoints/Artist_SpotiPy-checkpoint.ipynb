# SpotiPy 
## Top Artists streamed songs on Spotify from 2010 to 2019
### Most downloaded artists throughout 2010 to 2019.

```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import requests
import json
from scipy.stats import linregress
from sklearn.linear_model import LinearRegression 
```

```python
# Load the CSV file from Kaggle into Pandas DataFrame
top_artist = pd.read_csv("Resources/10years_top10s.csv")

top_artist
```

## Step 1: Filter the above dataframe for years 2010 - 2019

```python

# Clean grouped_data removing any columns that are too out of range to be useful.
top_artist.drop(['live', 'val', 'acous'], axis=1)

# Rename columns.  
top_artist = top_artist.rename(columns={'title': 'Title', 'artist': 'Artist', 'top genre': 'Genre', 'year': 'Year', 'bpm': 'BPM', 'nrgy': 'Energy', 'dnce': 'Dance', 'dB': 'Decibels (dB)', 'dur': 'Duration', 'spch': 'Speechiness', 'pop': 'Popularity'})

# Display the first 5 rows of the grouped dataframe
top_artist.head()
```

## Step 2: Create a bar graph for the top artist for each of the years from the filtered data.  Also, create and plot linear regression for each artist.

### Bar Graph for Top Artist in 2010-2019:

```python
# Get the top 10 most frequent artists in the DataFrame
top_artists = top_artist['Artist'].value_counts().nlargest(10)

# Create a bar plot of the top artists
plt.figure(figsize=(12, 6))
plt.bar(top_artists.index, top_artists.values, color='blue')

# Set the chart title and axis labels
plt.title('Top 10 Most Frequent Artists (2010-2019)')
plt.xlabel('Artist')
plt.ylabel('Frequency')

# Rotate the x-axis labels for readability
plt.xticks(rotation=45)

# Display the chart
plt.show()
```

```python
# Get the top 10 most frequent artists in the DataFrame
top_artists = top_artist['Artist'].value_counts().nlargest(10).index.tolist()

# Create a new DataFrame with the average values for each artist
artist_stats = top_artist.groupby('Artist').agg({'BPM': 'mean', 'Energy': 'mean', 'Dance': 'mean'}).loc[top_artists]

# Create a bar plot of the average BPM values for each artist
plt.figure(figsize=(12, 6))
plt.bar(artist_stats.index, artist_stats['BPM'], color='red', alpha=0.7)

# Set the chart title and axis labels
plt.title('Average BPM by Top 10 Artists')
plt.xlabel('Artist')
plt.ylabel('BPM')

# Rotate the x-axis labels for readability
plt.xticks(rotation=45)

# Display the chart
plt.show()
```

```python
# Create a bar plot of the average Dance values for each artist
plt.figure(figsize=(12, 6))
plt.bar(artist_stats.index, artist_stats['Dance'], color='green', alpha=0.7)

# Set the chart title and axis labels
plt.title('Average Dance by Top 10 Artists')
plt.xlabel('Artist')
plt.ylabel('Dance')

# Rotate the x-axis labels for readability
plt.xticks(rotation=45)

# Display the chart
plt.show()
```

```python
# Get the top 10 artists by frequency
top_artists = top_artist['Artist'].value_counts().head(10).index

# Create a horizontal bar chart of the Popularity scores for each of the top 10 artists
plt.figure(figsize=(8, 6))
plt.barh(top_artists, top_artist.groupby('Artist')['Popularity'].mean().loc[top_artists], color='orange')

# Set the chart title and axis labels
plt.title('Popularity Scores for Top 10 Artists')
plt.xlabel('Popularity')
plt.ylabel('Artist')

# Display the chart
plt.show()

```

```python
# Create a bar plot of the average Energy values for each artist
plt.figure(figsize=(12, 6))
plt.bar(artist_stats.index, artist_stats['Energy'], color='blue', alpha=0.7)

# Set the chart title and axis labels
plt.title('Average Energy by Top 10 Artists')
plt.xlabel('Artist')
plt.ylabel('Energy')

# Rotate the x-axis labels for readability
plt.xticks(rotation=45)

# Display the chart
plt.show()




```

```python
# Create a scatter plot of BPM vs. Popularity
plt.figure(figsize=(8, 6))
plt.scatter(top_artist['BPM'], top_artist['Popularity'], alpha=0.5)

# Set the chart title and axis labels
plt.title('BPM vs. Popularity')
plt.xlabel('BPM')
plt.ylabel('Popularity')

# Add a linear regression line to the chart
slope, intercept, r_value, p_value, std_err = linregress(top_artist['BPM'], top_artist['Popularity'])
x = np.linspace(top_artist['BPM'].min(), top_artist['BPM'].max(), 100)
y = slope * x + intercept
plt.plot(x, y, color='red')

if abs(r_value)>0.7:
    correlation_strength = 'Strong' 
else:
    correlation_strength = 'Weak'
legend_text = f'Correlation Strength: {correlation_strength}'
plt.legend([legend_text])

# Display the chart
plt.show()

```

```python
# Get the top 10 most frequent artists in the DataFrame
top_artists = top_artist['Artist'].value_counts().nlargest(10).index.tolist()

# Create a new DataFrame with the average values for each artist
artist_stats = top_artist.groupby('Artist').agg({'BPM': 'mean', 'Energy': 'mean', 'Dance': 'mean'}).loc[top_artists]

# Create a bar plot of the average values for each artist
plt.figure(figsize=(12, 6))
plt.bar(artist_stats.index, artist_stats['BPM'], color='red', alpha=0.7, label='BPM')
plt.bar(artist_stats.index, artist_stats['Energy'], color='blue', alpha=0.7, label='Energy')
plt.bar(artist_stats.index, artist_stats['Dance'], color='green', alpha=0.7, label='Dance')

# Set the chart title and axis labels
plt.title('Average BPM, Energy, and Dance by Top 10 Artists')
plt.xlabel('Artist')
plt.ylabel('Value')

# Rotate the x-axis labels for readability
plt.xticks(rotation=45)

# Add a legend and display the chart
plt.legend()
plt.show()
```
