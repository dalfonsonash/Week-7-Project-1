# SpotiPy 
## Top 100 streamed songs on Spotify from 2010 to 2019
### What music metrics set these songs apart from the rest.

```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import requests
import json
from scipy.stats import linregress
from sklearn.linear_model import LinearRegression 
```

```python
# Load the CSV file from Kaggle into Pandas DataFrame
top_songs = pd.read_csv("Resources/10years_top10s.csv")

top_songs
```

## Step 1: Filter the above dataframe for years 2010 - 2012

```python
top_songs = pd.read_csv("Resources/10years_top10s.csv")

# Clean grouped_data removing any columns that are too out of range to be useful.
top_songs.drop(['live', 'val', 'acous'], axis=1)

# Rename columns.  
top_songs = top_songs.rename(columns={'title': 'Title', 'artist': 'Artist', 'top genre': 'Genre', 'year': 'Year', 'bpm': 'BPM', 'nrgy': 'Energy', 'dnce': 'Dance', 'dB': 'Decibels (dB)', 'live': 'Live', 'val': 'Valence', 'dur': 'Duration', 'acous': 'Acousticness', 'spch': 'Speechiness', 'pop': 'Popularity'})

# Display the first 5 rows of the grouped dataframe
top_songs.head()
```

```python
# Filter the data for the years 2013, 2014, 2015
three_year_data = top_songs[top_songs['Year'].isin([2013, 2014, 2015])]
```

## Step 2: Create a bar graph for the top genres for each of the years from the filtered data.  Also, create and plot linear regression for each of the music metric columns.

### Bar Graph for Top Genres in 2010-2012:

```python
# Filter the data for year 2013
top_genres_2013 = three_year_data[three_year_data['Year'] == 2013]

# Get the value counts for the 'top genre' column
genre_counts = top_genres_2013['Genre'].value_counts()

# Create a bar graph
fig, ax = plt.subplots()
ax.bar(genre_counts.index, genre_counts.values)
ax.set_title('Top Genres for 2013')
ax.set_xlabel('Genre')
ax.set_ylabel('Number of Songs')
plt.xticks(rotation=70)

plt.show()
```

```python
# Filter the data for year 2014
top_genres_2014 = three_year_data[three_year_data['Year'] == 2014]

# Get the value counts for the 'top genre' column
genre_counts = top_genres_2014['Genre'].value_counts()

# Create a bar graph
fig, ax = plt.subplots()
ax.bar(genre_counts.index, genre_counts.values)
ax.set_title('Top Genres for 2014')
ax.set_xlabel('Genre')
ax.set_ylabel('Number of Songs')
plt.xticks(rotation=70)
plt.show()
```

```python
# Filter the data for year 2015
top_genres_2015 = three_year_data[three_year_data['Year'] == 2015]

# Get the value counts for the 'top genre' column
genre_counts = top_genres_2015['Genre'].value_counts()

# Create a bar graph
fig, ax = plt.subplots()
ax.bar(genre_counts.index, genre_counts.values)
ax.set_title('Top Genres for 2015')
ax.set_xlabel('Genre')
ax.set_ylabel('Number of Songs')
plt.xticks(rotation=70)
plt.show()
```

## Analyzing music metrics for the top genres from 2010 - 2012: 
### BPM metric for top genres 'Dance Pop', and 'Pop'. 

```python
# Filter data for the top genres 'dance pop' and 'pop' for years 2013, 2014, 2015
filtered_data = three_year_data[(three_year_data['Genre'].isin(['dance pop', 'pop'])) & (three_year_data['Year'].isin([2013, 2014, 2015]))]

# Group data by year and genre, and calculate the count
grouped_data = filtered_data.groupby(['Year', 'Genre']).size().unstack()

# Create bar plot
grouped_data.plot(kind='bar')

# Set axis labels and title
plt.xlabel('Year')
plt.ylabel('Number of Songs')
plt.title('Top Genres from 2013-2015')
plt.xticks(rotation=True)
# Set legend
plt.legend(title='Genre')

# Show plot
plt.show()




```

```python
# Calculate mean BPM per title and genre.

mean_bpm_per_title_genre = three_year_data.groupby(['Title', 'Genre'])['BPM'].mean().reset_index()

# Create bar graph for BPM per genre.
plt.figure(figsize=(12,6))
plt.bar(mean_bpm_per_title_genre['Genre'], mean_bpm_per_title_genre['BPM'])
plt.xlabel('Genre')
plt.ylabel('Average BPM')
plt.title('Average BPM per Song per Genre (2013-2015)')
plt.xticks(rotation=70)
plt.show()
```

```python
# Calculate mean Energy per title and genre.

mean_energy_per_title_genre = three_year_data.groupby(['Title', 'Genre'])['Energy'].mean().reset_index()

# Create bar graph for Energy per genre.
plt.figure(figsize=(12,6))
plt.bar(mean_energy_per_title_genre['Genre'], mean_energy_per_title_genre['Energy'])
plt.xlabel('Genre')
plt.ylabel('Average Energy')
plt.title('Average Energy per Song per Genre (2013-2015)')
plt.xticks(rotation=70)
plt.show()
```

```python
# Calculate mean Decibels per title and genre.

mean_db_per_title_genre = three_year_data.groupby(['Title', 'Genre'])['Decibels (dB)'].mean().reset_index()

# Create bar graph for Decibels per genre.
plt.figure(figsize=(12,6))
plt.bar(mean_db_per_title_genre['Genre'], mean_db_per_title_genre['Decibels (dB)'])
plt.xlabel('Genre')
plt.ylabel('Average dB')
plt.title('Average Decibels per Song per Genre (2013-2015)')
plt.xticks(rotation=70)
plt.gca().invert_yaxis()
plt.show()
```

```python
# Filter data for the top genres 'dance pop' and 'pop' for years 2013, 2014, and 2015
filtered_data = three_year_data[(three_year_data['Genre'].isin(['dance pop', 'pop'])) & (three_year_data['Year'].isin([2013, 2014, 2015]))]

# Create a box plot
plt.boxplot([filtered_data[filtered_data['Genre'] == 'dance pop']['BPM'],
             filtered_data[filtered_data['Genre'] == 'pop']['BPM']],
            labels=['dance pop', 'pop'])

# Set axis labels and title
plt.xlabel('Genre')
plt.ylabel('BPM')
plt.title('BPM Distribution by Genre (2013-2015)')

# Show plot
plt.show()

```

```python
# Filter data for 'dance pop' genre
dance_pop = three_year_data[three_year_data['Genre'] == 'dance pop']

# Create scatter plot
plt.scatter(x=dance_pop.index, y=dance_pop['BPM'], marker="o", edgecolors="Black")

# Set axis labels and title
plt.xlabel('Rank')
plt.ylabel('BPM')
plt.title('Dance Pop BPMs vs Rank 2013 - 2015')
plt.grid()

# Show plot
plt.show()
```

```python
# Subset the data for the dance pop genre
dance_pop = three_year_data[three_year_data['Genre'] == 'dance pop']

# Create a scatter plot of BPM vs Rank
plt.scatter(x=dance_pop.index, y=dance_pop['BPM'], marker="o", edgecolors="Black")

# Add axis labels and title
plt.xlabel('Rank')
plt.ylabel('BPM')
plt.title('BPM vs Rank 2013 - 2015')
plt.grid()

# Add a linear regression line
slope, intercept, r_value, p_value, std_err = linregress(dance_pop.index, dance_pop['BPM'])
plt.plot(dance_pop.index, intercept + slope*dance_pop.index, color='red', label='Correlation: Strong' if abs(r_value) >= 0.7 else 'Correlation: Weak')

# Add a legend
plt.legend()

# Display the plot
plt.show()
```

## Linear Regression models of music metrics for 2010.

```python
three_year_data.head(5)
```

```python
# Group by 'Title' and get the maximum 'Popularity' value
max_popularity = three_year_data.groupby(['Title', 'Genre'])['Popularity'].max()

# Create a new dataframe with the top 10 titles based on the highest popularity
top_10_titles = pd.DataFrame(max_popularity.nlargest(10))

# Reset the index of the dataframe
top_10_titles.reset_index(inplace=True)

# Print the top 10 titles dataframe
print(top_10_titles)
```

```python
# Combine 'Title' and 'Genre' for x-axis labels
x_labels = [f"{title}\n({genre})" for title, genre in zip(top_10_titles['Title'], top_10_titles['Genre'])]

# Create a bar graph of the top 10 titles
plt.figure(figsize=(8, 9))
plt.bar(x_labels, top_10_titles['Popularity'])
plt.xlabel('Title / Genre')
plt.ylabel('Popularity')
plt.title('Top 10 Titles by Popularity')
plt.xticks(rotation=70)
plt.tight_layout()
plt.show()



```

```python
# Select top 100 songs by rank in the year 2013
top_songs_2013 = three_year_data[three_year_data['Year'] == 2013].nlargest(100, 'Rank')
```

```python
# Create scatter plot
plt.scatter(x=top_songs_2013.index, y=top_songs_2013['BPM'])

# Add linear regression line
slope, intercept, r_value, p_value, std_err = linregress(top_songs_2013.index, top_songs_2013['BPM'])
plt.plot(top_songs_2013.index, intercept + slope*top_songs_2013.index, color='red')

# Set axis labels and title
plt.xlabel('Rank')
plt.ylabel('BPM')
plt.title('BPM vs Rank (Year 2013)')

# Add legend indicating correlation strength
if abs(r_value) > 0.7:
    correlation_strength = 'Strong'
else:
    correlation_strength = 'Weak'
legend_text = f'Correlation strength: {correlation_strength}'
plt.legend([legend_text])

# Show plot
plt.show()
```

```python
# Create scatter plot
plt.scatter(x=top_songs_2013.index, y=top_songs_2013['Energy'])

# Add linear regression line
slope, intercept, r_value, p_value, std_err = linregress(top_songs_2013.index, top_songs_2013['Energy'])
plt.plot(top_songs_2013.index, intercept + slope*top_songs_2013.index, color='red')

# Set axis labels and title
plt.xlabel('Rank')
plt.ylabel('Energy')
plt.title('Energy vs Rank (Year 2010)')

# Add legend indicating correlation strength
if abs(r_value) > 0.7:
    correlation_strength = 'Strong'
else:
    correlation_strength = 'Weak'
legend_text = f'Correlation strength: {correlation_strength}'
plt.legend([legend_text])

# Show plot
plt.show()
```
